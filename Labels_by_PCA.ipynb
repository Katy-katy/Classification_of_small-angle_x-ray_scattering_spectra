```python
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import sklearn
```

```python
import xrsdkit
from xrsdkit.tools.ymltools import read_local_dataset
```

```python
data = read_local_dataset('xrsdkit_modeling_dataset',downsampling_distance=1)
```

```python
from sklearn.decomposition import PCA

def doPCA(data, num_of_componets):
    pca = PCA(n_components=num_of_componets)
    pca.fit(data)
    return pca
```

```python
vis_data = data.copy()
```

```python
vis_data['dis'] = vis_data['system_class'].apply(lambda x: 1 if "disordered" in x  else 0)
```

```python
disordered=vis_data[vis_data.dis==1] 
non_disordered=vis_data[vis_data.dis==0]
```

```python
disordered.shape
```

```python
non_disordered.shape
```

```python
sep_data = pd.concat([disordered, non_disordered])
```

```python
from xrsdkit.tools.profiler import profile_keys
pca = doPCA(sep_data[profile_keys], 21)
transformed_data = pca.transform(sep_data[profile_keys])
```

```python
dis = transformed_data[ : 164]
non_dis = transformed_data[164: ]
```

### Disordered data by PCA components

```python
fig,axes =plt.subplots(7,3, figsize=(12, 9)) 

ax=axes.ravel()# flat axes with numpy ravel
for i in range(21):
  _,bins=np.histogram(transformed_data[:,i],bins=40)
  ax[i].hist(dis[:,i],bins=bins,color='r',alpha=.5)# red color for malignant class
  ax[i].hist(non_dis[:,i],bins=bins,color='g',alpha=0.3)# alpha is           for transparency in the overlapped region 
  t = "PCA component " + str(i)
  ax[i].set_title(t,fontsize=9)
  ax[i].axes.get_xaxis().set_visible(False) # the x-axis co-ordinates are not so useful, as we just want to look how well separated the histograms are
  ax[i].set_yticks(())
ax[0].legend(['true','false'],loc='best',fontsize=8)
plt.tight_layout()# let's make good plots
plt.show()
```
