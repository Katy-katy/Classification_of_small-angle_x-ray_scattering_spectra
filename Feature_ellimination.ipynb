```python
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
def make_plot(file_name):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    f = "feature_illim/"+ file_name
    f = open(f,'r')
    t = np.loadtxt(f,dtype=float,delimiter=',')
    y = t[ :, 0]
    x = t[ :, 1]
    ax.plot(x,y)
    ax.set_title(file_name)
```

```python
onlyfiles = [f for f in listdir("feature_illim") if isfile(join('feature_illim', f))]
```

```python
for f in onlyfiles:
    make_plot(f)
```
